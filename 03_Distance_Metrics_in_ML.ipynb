### **Distance Metrics** 
used in both supervised and unsupervised learning, **generally to calculate the similarity between data points.**

**Types of Distance Metrics in Machine Learning**
1. Euclidean Distance
2. Manhattan Distance
3. Minkowski Distance
4. Hamming Distance

**Few Machine learning algorithim uses Distance Metrics**
1. Clustering Algorithim
2. Classification - KNN Classification

#### **1. Euclidean Distance**
Euclidean Distance represents the **shortest distance between two points.**

Most machine learning algorithms including **K-Means use this distance metric to measure the similarity between observations.**

```python
# Importing necessary libraries 
from scipy.spatial import distance # To calculate distances
from google.colab import files
from IPython.display import Image
#uploaded = files.upload() # To import image from computer/tab
```

```python
print('Euclidean Distance Theory Overview : Calculate by Pythagorous Theorem\n')
Image('1A876F28-65F9-4243-B0C2-E4FBC2DDB56D.jpeg', width = 500)
```

```python
# Euclidean Distance
# defining the points
point_1 = (1, 2, 3)
point_2 = (4, 5, 6)

print('First Data Points :',point_1)
print('Second Data Points :', point_2)

#computing the euclidean distance
euclidean_distance = distance.euclidean(point_1, point_2)
print('\nEuclidean Distance b/w', point_1, 'and', point_2, 'is: ', euclidean_distance)
```

#### **2. Manhattan Distance**
##### Manhattan Distance is the sum of absolute differences between points across all the dimensions.

```python
print('Manhattan Distance vs Euclidean Distance')
Image('98236A32-D4FA-4359-B42C-377B289C90BD.png', width = 400)
```

```python
# Manhattan Distance vs Euclidean Distance
# defining the points
point_1 = (1, 1)
point_2 = (5, 4)

print('First Data Points :',point_1)
print('Second Data Points :', point_2)

#computing the euclidean distance
euclidean_distance = distance.euclidean(point_1, point_2)
print('\nEuclidean Distance b/w', point_1, 'and', point_2, 'is: ', euclidean_distance)

# computing the manhattan distance
manhattan_distance = distance.cityblock(point_1, point_2)
print('Manhattan Distance b/w', point_1, 'and', point_2, 'is: ', manhattan_distance)
```

#### **3. Minkowski Distance**

**Minkowski Distance is the generalized form of Euclidean and Manhattan Distance**
- If Lamda = 1, then it calculates Manhatten Distance
- If Lamda = 2, then it calculates Euclidean Distance

**In scipy package, the p parameter of the Minkowski Distance metric of SciPy package** 
1. When the order(p) = 1, it will represent Manhattan Distance 
2. When the order(p) = 2, it will represent Euclidean Distance.

```python
print('Manhattan Distance vs Euclidean Distance\n')
Image('11E4B729-8426-4200-92D6-C61D3307D082.png', width = 600)
```

```python
# Manhattan Distance vs Euclidean Distance
# defining the points
point_1 = (1, 1)
point_2 = (5, 4)

print('First Data Points :',point_1)
print('Second Data Points :', point_2)

#computing the euclidean distance
euclidean_distance = distance.euclidean(point_1, point_2)
print('\nEuclidean Distance b/w', point_1, 'and', point_2, 'is: ', euclidean_distance)

# computing the manhattan distance
manhattan_distance = distance.cityblock(point_1, point_2)
print('\nManhattan Distance b/w', point_1, 'and', point_2, 'is: ', manhattan_distance)

# computing the minkowski distance
minkowski_distance = distance.minkowski(point_1, point_2, p=1)
print('\nMinkowski Distance (p=1, Manhattan) b/w', point_1, 'and', point_2, 'is: ', minkowski_distance)

minkowski_distance = distance.minkowski(point_1, point_2, p=2)
print('\nMinkowski Distance (p=2, Euclidean) b/w', point_1, 'and', point_2, 'is: ', minkowski_distance)

minkowski_distance = distance.minkowski(point_1, point_2, p=3)
print('\nMinkowski Distance (p=3) b/w', point_1, 'and', point_2, 'is: ', minkowski_distance)
```

#### **3. Hamming Distance**
- Hamming Distance measures the **similarity between two strings of the same length** 

- The Hamming Distance **between two strings of the same length is the number of positions at which the corresponding characters are different.**

Lets say we have two strings:

**euclidean and manhattan**

- Since the length of these strings is equal, we can calculate the Hamming Distance. 
- We will go character by character and match the strings.
- The first character of both the strings (e and m respectively) is different. Similarly, the second character of both the strings (u and a) is different. and so on.
- **Look carefully**  seven characters are different whereas two characters (the last two characters) are similar:

**euclide--an and manhatt--an**

- Hence, the **Hamming Distance here will be 7.**
- **Note that larger the Hamming Distance between two strings, more dissimilar will be those strings (and vice versa)**



```python
# Hamming Distance
# defining two strings
string_1 = 'euclidean'
string_2 = 'manhattan'

# computing the hamming distance
hamming_distance = distance.hamming(list(string_1), list(string_2))*len(string_1)
print('Hamming Distance b/w', string_1, 'and', string_2, 'is: ', hamming_distance)
```
